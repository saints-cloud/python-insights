### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Passo a passo:
1. Importar a base de dados
2. Visualizar a base de dados
3. Tratamento de dados
4. Análise inicial -> quantos clientes cancelaram e qual o percentual de clientes
5. Análise da causa de cancelamento de clientes

```python
# Importar base de dados
import pandas as pd

tabela = pd.read_csv("cancelamentos_sample.csv")

# Deletando a coluna costumerID
tabela = tabela.drop(columns="CustomerID")

# Visualizar base de dados
display(tabela)
```

```python
# Tratamento de Dados -> Problemas: valores vazios e colunas úteis para se analisar
# Informações que não te ajudam, te atrabalham

# Excluir valores vazios
tabela = tabela.dropna()

# Verificação da tabela
display(tabela.info())

```

```python
# Análise inicial -> quantos clientes cancelaram e qual o percentual de cancelamento
# 0 para quem não cancelou e 1 para quem cancelou

display(tabela["cancelou"].value_counts())

# 1 -> 28393 e 0 -> 21603

# Percentual ou usar display(tabela["cancelou"].value_counts(normalize=True))
display(tabela["cancelou"].value_counts(1))
```
